```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('parliament-elections/parliament.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
print("Available columns:", df.columns.tolist())
print("Years available:", sorted(df['year'].unique()) if 'year' in df.columns else "Column 'year' not found")
print("Total rows:", len(df))
print("States:", df['state'].nunique() if 'state' in df.columns else "Column 'state' not found")
```

```python
# We only need 2004 onwards for our project
df_modern = df[df['YEAR'] >= 2004].copy()

print("Rows after 2004:", len(df_modern))
print("Years we have:", sorted(df_modern['YEAR'].unique()))
```

```python
# Check actual columns in df_modern
print("Actual columns in df_modern:", df_modern.columns.tolist())

# Identify the correct column names (try common variations)
constituency_col = [col for col in df_modern.columns if col.lower() in ['pcname', 'constituency', 'pc']][0] if any(col.lower() in ['pcname', 'constituency', 'pc'] for col in df_modern.columns) else None
year_col = [col for col in df_modern.columns if col.lower() == 'year'][0] if any(col.lower() == 'year' for col in df_modern.columns) else None
state_col = [col for col in df_modern.columns if col.lower() == 'state'][0] if any(col.lower() == 'state' for col in df_modern.columns) else None
votes_col = [col for col in df_modern.columns if col.lower() == 'votes'][0] if any(col.lower() == 'votes' for col in df_modern.columns) else None
party_col = [col for col in df_modern.columns if col.lower() == 'party'][0] if any(col.lower() == 'party' for col in df_modern.columns) else None

print(f"Using columns: year={year_col}, state={state_col}, constituency={constituency_col}, votes={votes_col}, party={party_col}")

# Winner = candidate with highest votes in each constituency per year
df_modern['WINNER'] = df_modern.groupby([year_col, state_col, constituency_col])[votes_col].transform('max') == df_modern[votes_col]

winners = df_modern[df_modern['WINNER'] == True].copy()
print("\nTotal winners:", len(winners))
print("Sample winners:")
print(winners[[year_col, state_col, constituency_col, party_col, votes_col]].head(10))
```

```python
party_seats = winners.groupby([year_col, party_col]).size().reset_index(name='SEATS')

# Top 5 parties overall
top5 = winners[party_col].value_counts().head(5).index.tolist()
print("Top 5 parties by seats won:", top5)

# Show seats per year for top parties
pivot = party_seats[party_seats[party_col].isin(top5)].pivot(index=year_col, columns=party_col, values='SEATS').fillna(0)
print("\nSeats per year:")
print(pivot)
```
